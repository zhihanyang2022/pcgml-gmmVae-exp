# Cornerplot-demo


```python
import json
import numpy as np
import corner
import matplotlib.pyplot as plt
```

```python
def load_json(fname:str)->np.array:
    with open(fname, 'r') as json_f:
        output_array = np.array(json.load(json_f))
    return output_array
```

```python
smb_metrics = load_json('smb_densities_difficulties.json')
smb_metrics = np.vstack([smb_metrics] * 100)[:10000]
ki_metrics = load_json('ki_densities_difficulties.json')
ki_metrics = np.vstack([ki_metrics] * 100)[:10000]
gen_metrics = load_json('sampled_densities_difficulties.json')
```

```python
# input arguments: samples from two distributions
fig = corner.corner(smb_metrics, color='red', plot_contours=True, labels=['Density', 'Difficulty'])
fig = corner.corner(ki_metrics, fig=fig, color='blue', plot_contours=True)
fig = corner.corner(gen_metrics, fig=fig, color='green', plot_contours=True)
# plt.tight_layout()
plt.savefig('cornerplot-demo.png', dpi=300)
```
